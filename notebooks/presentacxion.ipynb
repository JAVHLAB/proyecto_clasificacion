### **Problema de clasificación:** Clasificar las solicitudes de crédito en “aprobadas” o “rechazadas” a partir de variables demográficas, laborales y financieras del solicitante.

### **Algoritmo seleccionado:** Árboles de Decisión

Para el desarrollo del modelo de clasificación se seleccionó el algoritmo de Árboles de Decisión implementado mediante la librería scikit-learn
 en Python.

Un árbol de decisión es un algoritmo de aprendizaje supervisado utilizado para problemas de clasificación. Su funcionamiento consiste en dividir los datos en diferentes ramas según las características de entrada, hasta llegar a una decisión final o clase objetivo.

El árbol de decisión funciona mediante una estructura jerárquica compuesta por:

* Nodo raíz: representa el conjunto completo de datos.
* Nodos internos: contienen reglas o condiciones sobre los atributos.
* Ramas: representan el resultado de cada condición.
* Hojas: contienen la clasificación final.

```python
import joblib
import matplotlib.pyplot as plt
from sklearn.tree import plot_tree

model = joblib.load("../models/decision_tree_model.pkl")
columns = joblib.load("../models/columns.pkl")

plt.figure(figsize=(20,10))

plot_tree(
    model,
    feature_names=columns,
    class_names=["Rejected", "Approved"],
    filled=True,
    rounded=True,
    fontsize=8
)

plt.show()
```
