**<h1> Clasificación inteligente de datos </h1>**

### **Proyecto:** Sistema de clasificación para la aprobación de préstamos
### **Pregunta de investigación:** ¿La aprobación de un crédito depende de la edad, ingresos, monto del préstamo, puntaje crediticio, años de experiencia, género, nivel educativo, ciudad y tipo de empleo?
### **Hipótesis:** Existe un modelo que relaciona la edad, ingresos, monto del préstamo, puntaje crediticio, años de experiencia, género, nivel educativo, ciudad y tipo de empleo de una persona con si le será aprobado o no un crédito en una institución bancaria.
#### **Aprobación = edad + ingresos + monto_del_préstamo + puntaje_crediticio + género + nivel_educativo + ciudad + tipo_empleo**

### **Dataset:** loan_risk_prediction_dataset

### **Origen del dataset:** Fintech "Upstart" de Préstamos Personales y de Consolidación, recogidos entre enero y abril de 2022, con el objetivo de realizar valuación precisa del riesgo para minimizar pérdidas financieras y mejorar las decisiones de otorgamiento de préstamos. Este conjunto de datos simula perfiles realistas a partir de datos recolectados de prestatarios para ayudar a científicos de datos, analistas e investigadores a construir modelos predictivos para la aprobación de préstamos y la detección de incumplimientos.
